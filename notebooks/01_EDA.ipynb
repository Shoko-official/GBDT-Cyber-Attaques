# 📊 Analyse Exploratoire des Données (EDA) - NSL-KDD

Ce notebook explore le dataset NSL-KDD pour comprendre la structure des attaques et identifier les caractéristiques clés pour la modélisation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plots style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load processed data
df = pd.read_csv('../data/processed/train_processed.csv')
print(f"Dataset shape: {df.shape}")
```

## 1. Distribution de la Classe Cible (Binaire)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='target_binary', data=df, palette='viridis')
plt.title('Distribution: Normal (0) vs Attaque (1)')
plt.show()
```

## 2. Types d'Attaques (Multi-classe)

```python
plt.figure(figsize=(10, 6))
order = df['target_multi'].value_counts().index
sns.countplot(y='target_multi', data=df, order=order, palette='magma')
plt.title("Répartition des catégories d'attaques")
plt.show()
```

## 3. Corrélations avec la cible

Identifions les features les plus corrélées à la détection d'attaque (target_binary).

```python
correlations = df.corr()['target_binary'].sort_values(ascending=False)
print("Top 10 features corrélées positivement:")
print(correlations.head(11)) # including target itself
```

```python
top_features = correlations.abs().sort_values(ascending=False).index[1:11]
plt.figure(figsize=(12, 10))
sns.heatmap(df[top_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap des corrélations (Top features)')
plt.show()
```
